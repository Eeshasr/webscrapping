```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request  import urlopen
import logging
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "msi+laptop"
```

```python
flipkart_url
```

```python
urlclient = urlopen(flipkart_url)
```

```python
urlclient
```

```python
flipkart_page = urlclient.read()
```

```python
flipkart_page
```

```python
flipkart_html = bs(flipkart_page, 'html.parser')
```

```python
bigbox = flipkart_html.find_all('div' , {'class':"_1AtVbE col-12-12" })
```

```python
len(bigbox)
```

```python
del bigbox[0:3]
```

```python
productlink = 'https://www.flipkart.com'+bigbox[0].div.div.div.a['href']
```

```python
for i in bigbox :
    print('https://www.flipkart.com' + i.div.div.div.a['href'])
```

```python
product_req = requests.get(productlink)
```

```python
product_variable = bs(product_req.text , 'html.parser')
```

```python
comment_box = product_variable.find_all('div' , {'class': '_16PBlm'})
```

```python
len(comment_box)
```

```python
comment_box[0].div.div.find_all('p', {'class':'_2sc7ZR _2V5EHH'})[0].text
```

```python
user_names_list =[]
for i in comment_box:
    user_names = i.div.div.find_all('p', {'class':'_2sc7ZR _2V5EHH'})[0].text
    user_names_list.append(user_names)
    print(user_names)
```

```python
ratings_list =[]
for i in comment_box:
    ratings = i.div.div.div.div.text
    ratings_list.append(ratings)
    print(ratings)
```

```python
comment_summary_list =[]
for i in comment_box:
    comments_summery = i.div.div.div.p.text
    comment_summary_list.append( comments_summery)
    print (comments_summery)
```

```python
comments_list =[]
for i in comment_box:

    comments = i.div.div.find_all('div',{'class':''})[0].div.text
    comments_list.append(comments)
    print(comments)
```

```python
comment_box[0].div.div.find_all('div',{'class':''})[0].div.text
```

```python
import pandas as pd
```

```python
data_set = {'user_name':user_names_list , 'ratings': ratings_list , 'comments_summary':  comment_summary_list,'comments':comments_list}
```

```python
data = pd.DataFrame(data_set)
```
